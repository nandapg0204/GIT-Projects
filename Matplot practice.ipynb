```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(0,1,0.02)
```

```python
X
```

```python
y = X ** 5
```

```python
plt.plot(X,y,linewidth = 4,label = "X and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plotting between X and Y")
plt.legend()
plt.show()
plt.savefig("image.png")
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(X,y,label = "X and Y", )
ay = fig.add_axes([0.2,0.4,0.2,0.3])
ay.plot(y,X,label = "Y and X", )
```

```python
fig , axes = plt.subplots(1,2)
fig.set_size_inches(10,5)
axes[0].plot(X,y,label = "X and Y", color = "g", linewidth = 3)
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].set_title("Plotting X and Y")
axes[0].set_xlim(0,1)
axes[0].set_ylim(0,1)
axes[0].legend()



axes[1].plot(y,X , label = "Y and X", color = "b", linewidth = 3)
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")
axes[1].set_title("Plotting Y and X")
axes[1].set_xlim(0,1)
axes[1].set_ylim(0,1)
axes[1].legend()
plt.tight_layout()
```
